## 1) Conceitos Básicos da Estatística

Dados qualitativos: indicam uma qualidade, que não possibilita cálculos matemáticos. Dividem-se em nominais (sem ordenação) e ordinais (com ordenação): escolaridade, comida favorita, cor dos olhos, ser ou não fumante, estágio do câncer, etc...

Dados quantitativos: indicam uma quantidade, que possibilita cálculos matemáticos. Dividem-se em discretos (não contínuo) e contínuos: peso, altura, número de filhos, frequência de consumo de bebidas alcóolicas (em dias/semana), etc...

Para dados quantitativos, existe também outra classificação possível: se o nível é intervalar ou de razão, sendo a diferença que no último o "zero" representa "ausência" ou "nada" ou "nenhum". Exemplo: temperatura em °C (intervalar), temperatura em Kelvin (razão), distância de um centro de tratamento médico (razão), etc...

## 4) Tipos de Estudo e Planejamento

Estudo observacional: o pesquisador apenas se preocupa em coletar, organizar e analisar uma população sem intervenção.

Estudo experimental: o pesquisador divide a população em 2 ou mais grupos, a fim de observar como certas intervenções nas condições da população alteram os resultados obtidos. Para isso, sempre será usado um grupo de controle (grupo que não sofre intervenções).

Exemplos: teste de eficácia e de efeitos colaterais de um novo medicamento, pesquisa sobre satisfação, capacidade dos alunos quanto ao nível de escolaridade, etc...

Para estudos experimentais, pode ocorrer uma variável de confusão, na qual o pesquisador não sabe qual fator causou os efeitos provocados na variável de estudo. Também pode ocorrer o efeito placebo, no qual um indivíduo reage positivamente a um tratamento que, na realidade, não causa interferência nenhuma (placebo). A fim de evitar esse efeito, realiza-se o experimento cego, no qual o indivíduo não sabe se vai receber um tratamento ou placebo (chama-se de duplo-cego se o pesquisador também não sabe).

Pode haver também a separação dos indivíduos em diferentes grupos de tratamento, podendo ser feito de maneira totalmente aletória ou ser feito primeiramente agrupando indivíduos com características semelhantes e, então, fazer uma divisão aleatória.

Replicação: é a repetição de um experimento a fim de verificar sua validade.

## 5) Tipos de amostragem

Censo: medida de toda a população (Ex: Censo IBGE 20XX (https://censo2022.ibge.gov.br/)).

Amostragem: medida de parte de uma população (Ex: pesquisas quaisquer).

Observe os seguintes 5 tipos de amostragem:

1) Aleatória simples: qualquer elemento tem a mesma chance de ser selecionado.

2) Sistemática: cada elemento é associado a um número, que então são escolhidos em intervalos regulares (Ex: 7,12,17,22,27,...).

3) Estratificada: elementos de uma população são divididos em subconjuntos, de acordo com características chaves (Ex: idade, sexo, profissão, etc...). Na seleção de elementos, baseado nas proporções desses subconjuntos da população, um certo número de elementos é selecionado de cada subconjunto.

4) Por conglomerado: elementos de uma população são divididos em subconjuntos, mas todo subconjunto deve ser semelhante à população. Na seleção de elementos, um ou mais subconjuntos devem ser selecionados em sua íntegra.

5) Por conveniência: seleção dos elementos mais convenientes (mais acessíveis). Deve ser evitado.

## 1) Distribuições de frequência

Envolve na contagem de uma ou mais características dos dados de acordo com uma outra característica.

Ex: Consumo de bebida alcoólica em estudantes de acordo com a idade.

```python
# Exibição da "cabeça" do dataframe, assim como o maior e menor valor das idades
display(df.head())
i_max = df["age"].max()
print("max", i_max, sep = " = ")
print("min = ", end = "")
i_min = df["age"].min()
df["age"].min()
```

### a) Para dados quantitativos

Classe: subconjunto do grupo de dados criado a partir de um certo padrão (geralmente, divide-se os dados quantitativos em n grupos de mesmo tamanho em ordem crescente)

Ponto médio (marca de classe): a média aritmética dos limites da classe.

Frequência: contagem dos dados que aparecem na classe.

Frequência relativa: razão entre a frequência da classe e o número total de dados da
amostra.

Frequência acumulada: soma entre a frequência da classe e de todas as anteriores.


```python
# Criação da ditribuição de frequências das idades

distr_freq = df["age"].value_counts().sort_index()
display(distr_freq)
```

#### Histograma

É um grafico onde as frequências das classes são representadas pela altura de barras verticais distribuídas pelo eixo x.

```python
# Criação dos gráficos
# Amplitude é o tamanho de cada classe, podendo ser ou não um número inteiro

hist = df["age"].plot.hist(figsize = (5,3.75), color = 'blue')
hist.set_xlabel("Idade")
hist.set_ylabel("Frequência")
plt.show()
```

```python
hist3 = df["age"].plot.hist(bins=4, figsize = (4.97,3.75), grid = 'true', color= 'g')
hist3.set_xlabel("Idade")
hist3.set_ylabel("Frequência")
plt.xticks([x.left for x in pd.interval_range(start = i_min, end = i_max, freq = ((i_max - i_min) / 4))] + [i_max])
plt.show()
```

```python
bin_size = 1

hist6 = df["age"].plot.hist(bins = [i_min + i * round(bin_size) for i in range(int((i_max - i_min) / round(bin_size)) - 1) ] + [i_max],
                            figsize = (5,3.75), color = 'm', edgecolor = 'k', alpha = 0.7)
hist6.set_xlabel("Idade")
hist6.set_ylabel("Frequência")
plt.xticks([x.left for x in pd.interval_range(start = i_min, end = 21, freq = bin_size)] + [i_max])
barras = hist6.patches[:-1]
freqs = distr_freq

for barra, freq in zip(barras, freqs):
    altura = barra.get_height()
    hist6.text(x=barra.get_x()+barra.get_width()/2, y=altura+0.1, s=freq, ha='center', va='bottom')

hist6.text(x=hist6.patches[-1].get_x()+hist6.patches[-1].get_width()/2, y=hist6.patches[-1].get_height()+0.1,
           s=freqs[20]+freqs[21]+freqs[22], ha='center', va='bottom')

plt.show()
```

#### Gráfico de polígonos

Gráfico no qual as frequências das classes são pontos no gráfico de coordenadas (ponto médio da classe, frequência). Na construção desses gráficos, são inseridos dois pontos fictícios, um à esquerda e outro à direita dos pontos inicial e final, respectivamente, cada qual com frequência nula.

```python
pts_med = [m + 0.5 for m in distr_freq.index]
plot_df = pd.DataFrame(distr_freq.values, index=pts_med, columns=["Frequência"])
plot_df.loc[14.5] = 0
plot_df.loc[23.5] = 0
plot_df.sort_index(inplace=True)

plt.figure(figsize=(5, 3.75))
plt.plot(plot_df.index, plot_df["Frequência"], marker='o', linestyle='-', color='b')
plt.xlabel('Idades')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

#### Gráfico de ogiva

Gráfico no qual as frequências acumuladas das classes são pontos no gráfico de coordenadas (limite superior da classe, frequência acumulada). O gráfico se inicia com um ponto no limite inferior da primeira classe, de frequência acumulada nula.

```python
freqs_ac = distr_freq.cumsum()
freqs_ac = pd.concat([freqs_ac, pd.Series([0], index = [14])])

freqs_ac.sort_index(inplace=True)

plt.figure(figsize=(5, 3.75))
plt.plot(freqs_ac.index, freqs_ac.values, marker='o', linestyle='-', color='g')
plt.xlabel('Idade')
plt.ylabel('Frequência Acumulada')
plt.grid(True)
plt.show()
```

## 2) Medidas de tendência central

Valores que representam uma observação central de um conjunto de dados.

Média: é a média aritmética dos valores dos dados do conjunto de dados. Pode ser amostral ($\bar{x}$) ou populacional ($\mu$).

Mediana: é o valor que divide o conjunto de dados em dois (50% dos valores do conjunto estão acima dele, e os outros 50% abaixo). O valor pode ou não pertencer ao conjunto (no caso de um conjunto de dados com número par de dados, a mediana é definida como a média aritmética entre os dois valores "centrais").

Moda: é o valor mais comum do conjunto (o que mais aparece). Quando são $n$ valores que mais aparecem (todos esses $n$ valores possuem a mesma frequência), o conjunto é dito multimodal (no caso de $n=2$, o conjunto é dito bimodal).

Outliers: são valores do conjunto de dados que estão afastados da grande maioria dos dados. Ao se realizar os cálculos de medidas de tendência central (bem como outros também), é válida a retirada desses outliers para que os cálculos melhor representem o conjunto.

```python
plt.figure(figsize=(5, 3.75))
plt.scatter(df['absences'], df['G3'], alpha=0.5, color='blue')

plt.xlabel('Faltas')
plt.ylabel('Nota')
plt.title('Desempenho em função da presença')
plt.grid(True)

plt.show()

print("\nMédia = ", df['absences'].mean())
print("Mediana = ", df['absences'].median())
print("Moda = ", df['absences'].mode()[0])
```

## 3) Medidas de variação



Valores que representam a dispersão de um conjunto de dados.

Amplitude: é a diferena entre o menor e maior valor do conjunto ($A = x_{max} - x_{min}$).

Desvio: é a diferença entre o valor de um dado e a média do conjunto ($d = x - \bar{x}$ ou $d = x - \mu$).

Variância populacional: mede o quanto os dados estão afastados da média populacional; $\sigma^2 = \frac{\sum{(x-\mu)^2}}{N}$.

Desvio padrão populacional: é a raíz quadrada da variância; $\sigma = \sqrt{\frac{\sum{(x-\mu)^2}}{N}}$.

Variância amostral: $s^2 = \frac{\sum{(x-\bar{x})^2}}{n-1}$.

Desvio padrão amostral: $s = \sqrt{\frac{\sum{(x-\bar{x})^2}}{n-1}}$.

Coeficiente de variação: $CV = \frac{\sigma}{\mu}\cdot100\%$ ou $CV = \frac{s}{\bar{x}}\cdot100\%$

```python
plt.figure(figsize=(5, 3.75))
plt.scatter(df['absences'], df['G3'], alpha=0.5, color='blue')

plt.xlabel('Faltas')
plt.ylabel('Nota')
plt.title('Desempenho em função da presença')
plt.grid(True)

plt.show()

print(f"Amplitude = {df['absences'].max()-df['absences'].min()}")
print(f"Variância = {df['absences'].var()}")
print(f"Desvio padrão = {df['absences'].std()}")
print(f"CV = {round(df['absences'].std() / df['absences'].mean() * 100,2)}%")
```

## 4) Medidas de posição



São medidas que refletem a posição de um elemento no conjunto de dados.

Quantil: é um ponto que separa o conjunto de dados de tal forma que todos os valores inferiores a ele representam uma certa quantia do conjunto de dados.
Alguns quantis específicos:

Quartis: dividem o conjunto de dados em 4 partes;

Decis: dividem o conjunto de dados em 10 partes;

Percentis: dividem o conjunto de dados em 100 partes.

```python
plt.figure(figsize=(5, 3.75))
plt.scatter(df['absences'], df['G3'], alpha=0.5, color='blue')

plt.xlabel('Faltas')
plt.ylabel('Nota')
plt.title('Desempenho em função da presença')
plt.grid(True)

plt.show()

print(f" Percentil de 63% = {df['absences'].quantile(0.63)}")
print(f" 1° Quartil = {df['absences'].quantile(0.25)}")
print(f" 2° Quartil = {df['absences'].quantile(0.50)}")
print(f" 3° Quartil = {df['absences'].quantile(0.75)}")
```

Os quartis são comumente utilizados para a construção de boxplots:

```python
plt.figure(figsize=(6, 4))
plt.boxplot(df['absences'], vert=True)
plt.title('Faltas dos estudantes')
plt.show()

print("Tamanho do bigode = ", 1.5 * (df['absences'].quantile(0.75) - df['absences'].quantile(0.25)))
```
