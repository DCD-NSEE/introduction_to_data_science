### 3.2. Dicionários
Um dicionário é uma estrutura **realmente importante e útil**. Daqui em diante, classificaremos as estruturas usando as seguintes características:
1. **ordenada/não ordenada**: nos diz se a estrutura em questão **arranja seus dados** de de acordo com a ordem na qual foram declaradas.
2. **mutável/imutável**: caracteristica que nos diz se **podemos alterar os valores** dentro da estrutura deposi de já inseridos.
3. **indexada**: representa a organização da nossa estrutura, se ela representa um **index** qu nos permite acessar diretamente um valor

O dicionário no Python, por exemplo, é uma estrutura **não ordenada , mutável e indexada**.

Em Python, dicionários são escritos com chaves ```{}```, e eles possuem **chaves (keys)** e **valores (values)**. <br>
**Exemplo**:
```
population = {'SP': 45538936, 'RJ': 17159960, 'MG': 21040662, 'BA': 2}
```
Aqui, cada **UF** representa uma **chave** e sua **população** é associada ao **valor**. <br>
Agora, digamos que você queira saber a população de SP, então ```population['SP']``` lhe retornaria o resultado.

Obs.: Você pode, facilmente, **listar todas as chaves e valores** com os seguintes métodos built-in: **```population.keys()```, ```population.values()``` e ```population.items()```**

O método **items()** difere dos 2 anteriores, pois retorna dois valores iteráveis (que podem ser percorridos), um com as chaves e um com os valores do dicionário, sendo necessário atribuí-los à duas variáveis distintas quando utiizado em qualquer contexto.



````python
# Tente entender o código abaixo antes de rodar, e veja se o resultado faz sentido para você
dict_idades={'João':12,'André': 23,'Maria':22, 'Isabela':16}
for chave, valor in dict_idades.items():
    print(f'{chave} tem {valor} anos de idade')
````

### 3.4. Listas (Lists)
Listas também são uma estrutura **realmente importante e útil** em Python. Lsitas são estruturas **mutáveis, ordenadas e indexadas** Elas são criadas da mesma forma que tuplas, mas com colchetes, ```[]```. <br>
Listas também são indexadas e têm funções built-in similares a strings e tuplas.

Aqui, iremos aprender sobre **negative indexing** e **slicing**, mas esses conceitos também podem ser aplicados a strings e tuplas. Além disso, vamos ver uma forma muito elegante de se criar listas, através do que chamamos de **List Comprehension**.

#### 3.4.1. Básico

````python
# Criando uma lista
my_list = [5,1,7]
print(my_list)

# Adicionando elementos (método append)
my_list.append(2)
print(my_list)

# Removendo elementos
del(my_list[1])
print(my_list)

# Concatenação
my_list = my_list + [-1, -2, -5]
print(my_list)

# Extensão
my_second_list=[-10,-20]
my_list.extend(my_second_list)
print(my_list)

# in clause
print(2 in my_list)
````

#### 3.4.3. Slicing
Slicing é um **conceito muito importante** e nós vamos utilizá-lo bastante a partir de agora. É muito útil quando você quer fatiar sua sequência para **extrair elementos de certo intervalo** especificado pela seguinte sintaxe: 
<br><center>*(start index)* : *(end index)* : *(step)*</center> <br>
onde **start index está incluído** (***bolinha fechada***) e **end index está excluído** (***bolinha aberta***). Todos esses argumentos são opcionais e possuem valores *default*.

Execute a célula seguinte e veja os resultados:

````python
print("Lista original: ", my_list)

top5 = my_list[0:5] # O step default é 1
print("Primeiros 5 elementos da lista: ", top5)

odd_idx = my_list[1::2] # end index default é o último
print("Elementos em índices ímpares: ", odd_idx)

top7 = my_list[:7:1] # start index default é o primeiro
print("Elementos em índices ímpares: ", top7)


all_elements = my_list[::] 
print("Todos os elementos: ", all_elements)
````

#### 3.4.4. List comprehension
Uma forma bastante útil e elegante de construir uma lista é apresentada a seguir. Pode ser muito útil quando você quer **criar uma lista a partir de uma lista já existente**. <br>
Veja o exemplo abaixo e repare que **```range(10)```** cria uma lista de inteiros de 0 (incluído) a 10 (excluído).

````python
# Para cada elemento x no intervalo [0,10), faça 2^x
pow2 = [2**x for x in range(10)]
print(type(pow2))
print(pow2)
````

### 3.5. Sets
Sets são estruturas de dados importantes, análogas aos 'conjuntos' da matemática, porém, pouco abordadas em cursos de Python. Elas trazem muita similaridade com as tuplas, porém são declaradas usando chaves, ```{}```. <br>
Sets são estruturas de dados **não indexadas e não ordenadas**, porém são **mutáveis (valores ou tamanho)**. Um ponto muito importante e particularmente útil dos sets é que eles não aceitam valores duplicados e é possível fazer operações de conjuntos com os sets (união, interseção, disjunção, etc)

#### 3.5.1. Operações básicas
Os sets possuem operações básicas como ```.add()``` e ```.remove()```, similares às operações das listas, como mostra o exemplo abaixo:


````python
my_set={'A','B','C','D'} # criar um set
print(my_set)

my_set.add('D') # adicionar elementos ao set
my_set.add('E')
my_set.add('F')
print(my_set)

my_set.remove('A') # remover elementos do set
print(my_set)
````

Note que, mesmo tendo adicionado o valor ```D``` duas vezes, ele não foi duplicado no set.


Para entender uma das vantagens dos sets, vamos utilizar uma lista de valores duplicados e comparar o tempo necessário para encontrar os valores únicos dessa lista com uma função e transformando-a em um set.

O exemplo abaixo constrói uma função para encontrar os valores únicos de uma lista. Observe a diferença no tempo decorrido quando utilizamos a função definida e o set.

````python
def get_unique_values(any_list):
    unique_values=[]
    for value in any_list:
      if value not in unique_values:
          unique_values.append(value)
    return unique_values

import time # Importa o modulo de tempo
my_list=['a','b','c']*100000 # cria uma lista com 300 mil itens
start = time.time()
print(get_unique_values(my_list)) # usa a função criada
end = time.time()
print(f'Tempo necessário= {end-start}')

start = time.time()
print(set(my_list)) # usa um set
end = time.time()
print(f'Tempo necessário= {end-start}') 
````

Observe que a operação utilizando o set foi aproximadamente 7 vezes mais rápida!
